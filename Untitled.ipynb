```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = [250/128,250/64,250/32,250/16,250/8,250/4,250/2,250]
y = [-0.03894, -0.03304, 0.0209, 0.02095, 0.1125, 0.2412, 0.4898, 1.415] 
x_regress = [5.959, 7.006, 16.57, 16.58, 32.82, 55.64, 99.74, 263.7]
fig = plt.figure(figsize = (7, 5))
rat = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(rat) 
# poly1d_fn is now a function which takes in x and returns an estimate for y
print(poly1d_fn)
fig, ax = plt.subplots()
ax.plot(x,y, 'o', x, poly1d_fn(x), 'k') 
ax.plot(x_regress,y, 'oy')
plt.text(.125, .75, '$y = 0.005638x - 0.072537$ \n$R^2 = 0.98162239$', ha='left', va='bottom',transform=ax.transAxes)
ax.set_title('Reported Optical Density (OD) as a function of IL-8 concentration')
ax.set_xlabel('IL-8 concentration (pg/ml)')
ax.set_ylabel('Optical density')
plt.show()
```

```python
# import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
import tkinter as tk
from tkinter import filedialog
import scipy.stats as stats

root = tk.Tk()
root.withdraw()

#receive input
file_path = filedialog.askopenfilename()
col1 = input('column 1 name')
col2 = input('column 2 name')
ax_title = input('Axis title')
ax_ylabel = input('Y-axis label')
file_name = input('Name of the file (.jpg)') + '.jpg'
#read .csv file as dataframe using specific column
df = pd.read_csv(file_path, usecols=[col1,col2])

print(df)

#swap column to make sure without saponin comes first
#df= df[['without saponin', 'with saponin']]

#make list from DataFrame column
cols = list(df)

#calculate means, standard deviation, and unpaired t-test (man-whitney u test)
means = df.mean(axis = 0).to_list()
std = df.std(axis = 0).tolist()
stat, p_value = stats.mannwhitneyu(x=df[col1], y=df[col2], alternative = 'two-sided', method = 'exact')

#melt dataframe to make the value labeled with column
df = pd.melt(frame = df, var_name = 'variable', value_name = 'value')

#start subplot
fig, ax = plt.subplots()

ax.errorbar(cols, means,
            yerr=std,
            marker = 'o', linestyle = 'None', color = 'k', ecolor = 'k', elinewidth = 1, capsize = 4)           

ax = sns.swarmplot(ax = ax,
              data = df,
              x = 'variable',
              y = 'value', hue = 'variable')

#remove legend box
ax.legend([],[], frameon=False)

ax.set_ylabel(ax_ylabel)
ax.set_xlabel('')
ax.set_title(ax_title)

# ha and va tells where axis goes positive (e.g. right top means 0,0 at left buttom and goes positive at right and top - Quartile 1)
plt.text(.99, .99, '$P = $'+ str(round(p_value,3)), ha='right', va='top', transform=ax.transAxes)
plt.savefig(file_name, dpi=120)
plt.show()
```
